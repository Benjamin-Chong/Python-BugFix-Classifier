## 2.1 Creating the splits for V1 and 
1. Load the cleaned dataset
2. Creates the train/test split
3. Saves the train and test split into artifacts for V1 and V2

```python
from sklearn.model_selection import train_test_split
from pathlib import Path
import pandas as pd

df = pd.read_csv('../../data/processed/cleaned_v1.csv')
X, y = df['diff'], df['top_level_label']

X_train_raw, X_test_raw, y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 42, stratify=y)

Path("../../data/processed").mkdir(parents=True, exist_ok=True)
train = pd.concat([X_train_raw, y_train], axis=1).reset_index(drop=True)
test = pd.concat([X_test_raw, y_test], axis=1).reset_index(drop=True)

train.to_csv('../../data/processed/train_v1.csv', index=False)
test.to_csv('../../data/processed/test.csv', index=False)
```

## 2.2 Verifying Shape and Empty Rows

```python
assert X_train_raw.index.equals(y_train.index)
assert X_test_raw.index.equals(y_test.index)

print(f'Train Shape: {train.shape}')
print(f'Test Shape: {test.shape}')
print(f'Train Columns: {train.columns}')
print(f'Test Columns: {test.columns}')
print('Train Set:')
print(train.isna().sum())
print('Test Set:')
print(test.isna().sum())
```
