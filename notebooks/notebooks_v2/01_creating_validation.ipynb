## 1 Generating the Validation Set

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from pathlib import Path

train = pd.read_csv('../../data/processed/train_v1.csv')

train_v2, validation_v2 = train_test_split(train, test_size=.20, random_state=42, stratify=train['top_level_label'])

train_v2 = train_v2.reset_index(drop=True)
validation_v2 = validation_v2.reset_index(drop=True)

Path('../../data/processed').mkdir(parents=True, exist_ok=True)

train_v2.to_csv('../../data/processed/train_v2.csv', index=False)
validation_v2.to_csv('../../data/processed/validation_v2.csv', index=False)
```
